```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(7,3))

w0 = 20
t = np.linspace(0,10,500)

def omega(t): return w0 - 8*np.exp(-(t-2.5)**2/2) - 3*np.exp(-(t-7.5)**2/2)
#plt.plot(t, omega(t)/w0)
def y(t): return np.sin(omega(t)*t)
plt.plot(t, y(t)); plt.xticks([]); plt.yticks([]); plt.grid()
```
