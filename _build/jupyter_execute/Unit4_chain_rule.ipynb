# Chain Rule

## Composite functions
A composite function is a function of another function, i.e., 
\begin{align}
y = f(g(x)) \stackrel{\text{or}}{=} (f\circ g)(x),
\end{align}
which may be expressed using an intermediate variable $v$:
\begin{align}
v &= g(x)\nonumber\\
y &= f(v)\nonumber
\end{align}
or diagrammatically

$$
x \rightarrow\boxed{g}\stackrel{v}{\longrightarrow}\boxed{f}\rightarrow y.
$$

Similarly, given $f(x)$, $g(x)$ and $h(x)$, we can form a composite function
\begin{align}
y = f(g(h(x))) \stackrel{\text{or}}{=} (f\circ g\circ h)(x),
\end{align}
which means
\begin{align}
v&=h(x)\nonumber\\
w &= g(v)\nonumber\\
y &= f(w)\nonumber
\end{align}
or

$$
x\rightarrow\boxed{h}\stackrel{v}{\longrightarrow} \boxed{g} \stackrel{w}{\longrightarrow} \boxed{f} \rightarrow y.
$$

## Chain rule
Let us expand a composite function $y=f(g(x))$ by substituting $x+dx$:
\begin{align}
y(x+dx) =  f(g(x+dx)) = f( g(x) + g'(x) dx).
\end{align}
Since $\ds g' dx = \frac{dg}{dx}\cdot dx = dg$,
\begin{align}
y(x+dx) = f(g + dg) = f(g) + f'(g)\,dg.
\end{align}
Substituting $dg = g' dx$ back in the expression, we have
\begin{align}
y(x+dx) = f(g(x)) + f'(g(x)) g'(x) \,dx.
\end{align}
Since the coefficient of $dx$ is the derivative,
\begin{align}
y'(x) = \frac{d}{dx} f(g(x)) =f'(g(x))\cdot  g'(x) \stackrel{\text{or}}{=} \frac{df}{dg}\cdot \frac{dg}{dx}.
\end{align}

## Triple chain rule and more
The chain rule works in a similar way for composite functions made up of 3 or more functions.  For example, the derivative of $y(x) = f(g(h(x)))\stackrel{\text{or}}{=} (f\circ g\circ h)(x)$ is
\begin{align}
y'(x) = \frac{d}{dx} f(g(h(x))) =f'(g(h(x)))\cdot g'(h(x))\cdot h'(x) \stackrel{\text{or}}{=} \frac{df}{dg}\cdot \frac{dg}{dh}\cdot \frac{dh}{dx}.
\end{align}

## Related Rates
Consider the following situation.
The quantity $f$ is a function of another quantity $g$, and $g$ in turn is a function of time $t$. Then using the chain rule, the rate of change in $f$ in time is
\begin{align}
\frac{d}{dt}f(g(t)) = f'(g(t))\cdot g'(t) = \frac{df}{dg}\cdot \frac{dg}{dt}
\end{align}

```{admonition} Answer

```

## Forward mode automatic differentiation

As we have seen in the previous units, there are different ways to differentiate a function. Let us review how we compute 
the derivative of $f(x) = \sin^2(x)$ at $x=\pi/4$ in these two ways:

1. Using the definition of the derivative, we can calculate

$$f'\left(\frac{\pi}{4}\right)\approx \frac{\sin^2(\frac{\pi}{4}+0.01) - \sin^2(\frac{\pi}{4})}{0.01} = 0.99993.$$

2. Computing the derivative symbolically
\footnote{You could use the trigonometric identity $2\sin(x)\cos(x)=\sin(2x)$.}, we have

$$\frac{d}{dx}\sin^2(x) = 2\sin(x)\cos(x) ~~\therefore f'\left(\frac{\pi}{4}\right) = 2\sin\left(\frac{\pi}{4}\right)\cos\left(\frac{\pi}{4}\right) = 1.$$

There is another way, namely, automatic differentiation. Although it is a numerical method, it computes the derivative **exactly** (up to the precision of the computer).

Let $f$ be a function of $v$, and $v$ be a function of $x$.  We are interested in computing $\frac{df}{dx}$ in the end.   
We have

$$f(v(x+dx)) = f\left(v(x)+\frac{dv}{dx}(x)\cdot dx \right) = f(v) + f'(v)\cdot\frac{dv}{dx}\cdot dx$$

We now express $f(v)$ using *dual numbers*:

$$f(v) \to  f(v) + f'(v)\cdot\frac{dv}{dx}\cdot dx \to \left(f(v),~f'(v)\cdot \frac{dv}{dx}\right),
$$(dualfunction)

where the derivative $\frac{df}{dx} =f'(v)\cdot \frac{dv}{dx}$ appears in the second component of the dual number.
In general, a dual number has the form

$$u = (u_f,~ u_d),$$

where $u_f$ is the value of the function $u$, and $u_d$ is the value of the derivative $du/dx$.
In particular, for the variable, we are differentiating with ($x$) and any constant ($c$) are represented in dual number form we have

$$x = (x_f, 1) ~~ \text{and}~~ c = (c_f, 0).$$

Also, we define elementary functions and operations
 in dual numbers in the form of {eq}`dualfunction`:
\begin{align}
c\cdot u &= (c\cdot u_f,~ c\cdot u_d) &\because 
	(c\cdot u)' = c\cdot u'~~~~~~~~~~~~~~~~~\\
 u \pm w &= (u_f\pm w_f,~ u_d\pm w_d) &\because
 	(u \pm w)' = u' \pm w'~~~~~~~~~\\
 u\cdot w &= (u_f\cdot w_f,~ u_d\cdot w_f + u_f\cdot w_d) &\because 
	(u\cdot w)' = u'\cdot w + u\cdot w'\\
 \frac{u}{w} &= \left( \frac{u_f}{w_f},~  \frac{u_d\cdot w_f - u_f\cdot w_d}{w_f^2}\right) &\because 
 	\left( \frac{u}{w} \right)' = \frac{u'\cdot w - u\cdot w'}{w^2}\\
u^n &= ( u_f^n, ~ n\cdot u_f^{n-1}\cdot u_d)&\because
 	(u^n)' = n u^{n-1}~~~~~~~~~~~~~~~~\\
\sin(u) &= \left(\sin(u_f),~ \cos(u_f)\cdot u_d\right)&\because
 	\sin(u)' = \cos(u)~~~~~~~~~~~~~\\
 e^u &= \left(e^{u_f},~ e^{u_f}\cdot u_d\right)&\because
 	(e^u)' = e^u~~~~~~~~~~~~~~~~~~~~~~~\\
 \vdots\nonumber
\end{align}
Here is an example of how it can be implemented in Python:

````python
import numpy as np

class dual:
    def __init__(self, f,d):
        self.f = f
        self.d = d
    def __add__(self, other):
        return dual(self.f + other.f, self.d + other.d)
    def __sub__(self, other):
        return dual(self.f - other.f, self.d - other.d)
    def __neg__(self):
        return dual(-self.f, -self.d)
    def __mul__(self, other):
        return dual(self.f * other.f, self.f * other.d + self.d * other.f)
    def __truediv__(self, other):
        return dual(self.f / other.f, (self.d * other.f - self.f * other.d)/other.f**2)
    def __pow__(self, num):
        return dual(self.f**num, num*self.f**(num-1)*self.d)
    def sin(self):
        return dual(np.sin(self.f), np.cos(self.f)*self.d)
    def cos(self):
        return dual(np.cos(self.f), -np.sin(self.f)*self.d)
    def exp(self):
        return dual(np.exp(self.f), np.exp(self.f)*self.d)
    def log(self):
        return dual(np.log(self.f), self.d/self.f)
    def sqrt(self):
        return dual(np.sqrt(self.f), 1/2/np.sqrt(self.f) * self.d)

#Demonstration of forward-mode automatic differentiation
def exp(x): return x.exp()
def f(x):
    if x.f > 0:
        return x**2 * exp( -(x-dual(3,0))**2 / dual(2,0) ) # f = x^2 exp(-(x-3)^2 /2)
    else:
        return -x**2 * exp( -(x+dual(3,0))**2 / dual(2,0) )# f = - x^2 exp(-(x+3)^2 /2)

import matplotlib.pyplot as plt
x = np.linspace(-7,7,200)
plt.figure(figsize=(5,2)); plt.xlim([-7,7]); plt.ylim([-12,12])
plt.plot(x, [f(dual(xv,1)).f for xv in x], label="f(x)")
plt.plot(x, [f(dual(xv,1)).d for xv in x], label="f'(x)")
plt.plot([-8,8],[0,0], [0,0], [-12,12], lw=0.5, color="k")
plt.legend(); plt.show()
````

Now, let us compute the derivative of $f(x) = \sin^2(x)$ with respect to $x$ evaluated at $x=\frac{\pi}{4}$ using dual numbers.
First, 
Now, the derivative of $x$ with respect to $x$ is 1, so we write

$$x = \left(\frac{\pi}{4},~ 1\right)$$

Then,

$$v = \sin(x) = (\sin(x_f),~ \cos(x_f)\cdot x_d) = \left(\sin\left(\frac{\pi}{4}\right), ~ \cos\left(\frac{\pi}{4}\right)\cdot 1\right) = \left(\frac{1}{\sqrt{2}},~ \frac{1}{\sqrt{2}}\right)$$

and finally,

$$f(x) = \sin^2(x) = v^2 = (v_f^2,~ 2v_f\cdot v_d) = \left(\left(\frac{1}{\sqrt 2}\right)^2,~ 2\cdot\frac{1}{\sqrt 2}\cdot\frac{1}{\sqrt 2}\right) = \left(\frac{1}{2},~ 1\right),$$

which states

$$f\left(\frac{\pi}{4}\right) = \frac{1}{2}~~ \text{and} ~~ f'\left(\frac{\pi}{4}\right) = 1.$$

This procedure is called the *forward-mode automatic differentiation*.
